```python
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
```

```python
def image_enhance(fn, bright, contr):
    path = '../figures/submission-figures/' + fn + '.tiff'
    img = Image.open(path).convert('L')
    img = ImageEnhance.Brightness(img).enhance(bright)
    img = ImageEnhance.Contrast(img).enhance(contr)
    fig, ax = plt.subplots(dpi=300)
    ax.imshow(img, cmap='gray')
    img.save('../figures/figures-grayscale/' + fn + '.tiff')
    img.save('../figures/figures-grayscale/' + fn + '.png')
```

```python
image_enhance('kase-glomb_fig1', 2, 2)
```

```python
image_enhance('kase-glomb_fig2', 2, 2)
```

```python
  # cmap='gray' specifies to plt.imshow() that it should use the grayscale color map
image_enhance('kase-glomb_fig3', 1, 1)
```

```python
image_enhance('kase-glomb_fig5', 1, 1.2)
```

```python
image_enhance('kase-glomb_fig6', .5, 1.3)
```
